```python
#INSTRUCTIVO
#Cada escenario esta distribuido por diferentes tipos de minerales
#Estarán distribuidos por dureza y densidad

#El número 1 representa Verdadero y el número 0 representa Falso

def AgenteClasificador():
    '''
    Función que determina el agente 
        parametros:
            none
        returns:
            none
    '''
    
    costo = 0
    mineral = 'Piedra'
    #Ingreso de piedra
    while True:
        try:
            ingresoPiedra = int(input('Ingresa una piedra? 1/0: '))
            if(ingresoPiedra == 1 or ingresoPiedra == 0):
                print('Se ha recibido un registro')
                break;
            else:
                print('Solo se ingresa 1 ó 0')
        except:
            print('No se ha ingresado un número')
    
    
    if(ingresoPiedra == 1):
        #Ingreso de la densidad
        while True:
            try:
                ingresoDensidad = float(input('Ingrese la densidad de la piedra (g/cm3): '))
                if(ingresoDensidad > 1.5 or ingresoDensidad < 6):
                    print('Densidad ingresada')
                    break;
                else:
                    print('Ingrese un dato correcto')
            except:
                print('No se ha ingresado un dato numérico')
    
        #Ingreso de la dureza
        while True:
            try:
                ingresoDureza = float(input('Ingresa la dureza de la piedra (1-10): '))
                if(ingresoDureza >= 1 or ingresoDureza <= 9):
                    print('Se ha ingresado una piedra')
                    break;
                else:
                    print('El rango es entre 1 y 10')
            except:
                print('No se ha ingresado un dato numérico')
        
        #Se plantean los escenarios para el procesamiento
        #Cada escenario cuenta con una estimación de dureza y densidad establecidas como rango
        
        #Para Gypsum y Halita
        #~Rango dureza 2-2.5
        
        if(ingresoDureza >= 2 and ingresoDureza <= 2.5):
            if(ingresoDensidad >= 2 and ingresoDensidad <= 2.165):
                mineral = 'Halita'
            else:
                print('No es halita')
            
            if(ingresoDensidad >= 2.3 and ingresoDensidad <= 2.33):
                mineral = 'Gypsum'
            else:
                print('No es Gypsum')
        
        
        
        #Para la Calcita
        #Rango dureza 3-3.5
        if(ingresoDureza >= 3 and ingresoDureza <= 3.5):
            if(ingresoDensidad >= 2.7 and ingresoDensidad <= 2.72):
                mineral = 'Calcita'
            else:
                print('No es calcita')
        
        
        #Para la Dolomita y Aragonita
        #Rango dureza 3.5-4
        if(ingresoDureza >= 3.5 and ingresoDureza <= 4):
            if(ingresoDensidad >= 2.8 and ingresoDensidad <= 2.84):
                mineral = 'Dolomita'
            else:
                print('No es dolomita')
            
            if(ingresoDensidad >= 2.9 and ingresoDensidad <= 2.94):
                mineral = 'Aragonita'
            else:
                print('No es aragonita')
        
        #Para Fluorita
        #Rango dureza 4 y 4.5
        if(ingresoDureza >= 4 and ingresoDureza <= 4.5):
            if(ingresoDensidad >= 3 and ingresoDensidad <= 3.5):
                mineral = 'Fluorita'
            else:
                print('No es fluorita')
                
        
        
        #Para la Magnetita
        #Rango dureza 5-6
        if(ingresoDureza >= 5 and ingresoDureza <= 6):
            if(ingresoDensidad >= 5.1 and ingresoDensidad <= 5.2):
                mineral = 'Magnetita'
            else:
                print('No es magnetita')
        
        #Para Cuarzo
        #Rango dureza 7-7.5
        if(ingresoDureza >= 7 and ingresoDureza <= 7.5):
            if(ingresoDensidad >= 2.6 and ingresoDensidad <= 2.66):
                mineral = 'Cuarzo'
            else:
                print('No es cuarzo')
        
        
        #Se muestra por pantalla el mineral encontrado
        print('El mineral encontrado es: ',mineral)
            
        
        
    else:
        print("No se ha ingresado ninguna piedra")
    
    
    
        

AgenteClasificador()
        
    
    
```
